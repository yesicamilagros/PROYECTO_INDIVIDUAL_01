# valores nulos de los campos revenue y budget : rellenando con 0

```python
# modificacion del tipo de dato en revenue y budget
movie.loc[[19725,29491,35575]]
```

```python
# de las filas anteriores se observa que en las filas existe datos que no guardan relacion por tanto se ha determinado eliminarlas del dataframe

movie=movie.drop([19725,29491,35575])
```

```python
# reindexamos
movie=movie.reset_index(drop=True)
```

```python
# modificacion de tipo de dato y llenado de valores vacios y nulos en revenue y budget
movie['revenue']=movie['revenue'].astype(float)
movie['budget']=movie['budget'].astype(float)
movie['revenue']=movie['revenue'].fillna(0)
movie['budget']=movie['budget'].fillna(0)
```

# tratamiento de valores nulos en release_date

```python
#se eliminan los valores nulos de release date ya que release_date se utilizara en las funciones de consulta las fechas de lamzamiento son claves para una mejor interpretacion de comparacion

movie=movie.dropna(subset=['release_date']).reset_index()
```

# conversion de la columna release date

```python
#conversion de la columna release date o fecha de estreno   de str a datetime

movie['release_date']=movie['release_date'].apply(lambda x : datetime.strptime(x, '%Y-%m-%d'))
```

# creacion de la columna release_date

```python
# generamos la columna year o año de estreno

movie['Year'] = movie['release_date'].dt.year
```

```python
# renombramos la columna year por release_year

movie.rename(columns={"Year": "release_year"},inplace=True)
```

## creacion de la columna retorno de inversion (return)

```python
lista=[]
for i in range(movie.shape[0]):
  if movie['budget'][i]!=0:
     lista.append(movie['revenue'][i]/movie['budget'][i])
  else:
     lista.append(0.0)
movie['return']=pd.Series(lista)
```
